## Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
DATA_PATH = "datasets/"
NETFLIX_COLOR = "#b20710"
```

```python
df = pd.read_csv(DATA_PATH + "netflix_titles.csv")
df.head()
```

## Check For Missing Values

```python
# Checks which columns have NaNs
col_hasnans = [(col, df[col].isnull().any()) for col in df.columns]
nan_cols = [pair[0] for pair in col_hasnans if pair[1] == True]

print("Columns that have NaNs:")
print(nan_cols)
```

```python
# Percentages of NaNs
nan_percentages = [(df[col].isnull().sum() / len(df[col]) * 100) for col in nan_cols]

plt.title("Percentages of NaNs in each column")
plt.bar(nan_cols, nan_percentages, color=NETFLIX_COLOR)
plt.ylabel("(%)")
plt.xlabel("Columns")
```

## Convert date_added column to datetime

```python
# For any conversion errors, the value will be NaT
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

df.info()
```

## Movies vs TV Shows

```python
# Count of movies and TV shows
num_movies, num_tvshows = len(df[df["type"] == "Movie"]), len(df[df["type"] == "TV Show"])


percentages = [(num_movies / len(df) * 100), (num_tvshows / len(df) * 100)]
percentages = [f"{p:.2f}%" for p in percentages]

plot = plt.bar(("Movies", "TV Shows"), [num_movies, num_tvshows], color=NETFLIX_COLOR)
plt.title("Movies vs TV Shows")
plt.ylabel("Count")
plt.bar_label(plot, percentages)  # Adds percentages as labels to plot
plt.show()
```

## Distribution of countries

```python
"""
Create dictionary of each unique country and their total count in the dataset
"""
country_counts_dict = {}  # Stores the total counts of each country

country_counts_series = df["country"].value_counts()

for countries in country_counts_series.index:
    list_of_countries = countries.split(",")

    for c in list_of_countries:
        if len(c) > 0:
            c = c.strip()  # Remove whitespace
            value = country_counts_series[countries]
            
            
            # Check if country is already in dictionary, 
            # then increment the value already there, else set a new value
            if c in country_counts_dict:
                country_counts_dict[c] += value
            else:
                country_counts_dict[c] = value
```

```python
# Remove countries from dictionary with small counts
data = {k: v for k, v in country_counts_dict.items() if v > 25}
data = pd.Series(data).sort_values(ascending=False)


plt.figure(figsize=(16, 9))
plt.bar(data.index, data.values, color=NETFLIX_COLOR)
plt.title("Distribution of Countries")
plt.xlabel("Countries")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.plot()
```

## Movies vs TV Shows Added by Release Year

```python
movies_year_counts = df.loc[df["type"] == "Movie", "release_year"].value_counts()
tvshows_year_counts = df.loc[df["type"] == "TV Show", "release_year"].value_counts()

plt.figure(figsize=(16, 9))
plt.bar(movies_year_counts.index, movies_year_counts.values, label="Movies", color=NETFLIX_COLOR)
plt.bar(tvshows_year_counts.index, tvshows_year_counts.values, label="TV Shows", color="#333")
plt.title("Movies vs TV Shows Added by Release Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
```

## Distribution by Genres

```python
"""
Create dictionary of each unique genre and their total count in the dataset
"""
genre_counts_dict = {}  # Stores the total counts of each country

genre_counts_series = df["listed_in"].value_counts()

for genres in genre_counts_series.index:
    list_of_genres = genres.split(",")

    for g in list_of_genres:
        if len(g) > 0:
            g = g.strip()  # Remove whitespace
            value = genre_counts_series[genres]
            
            
            # Check if genre is already in dictionary, 
            # then increment the value already there, else set a new value
            if g in genre_counts_dict:
                genre_counts_dict[g] += value
            else:
                genre_counts_dict[g] = value
```

```python
data = pd.Series(genre_counts_dict).sort_values(ascending=False)

plt.figure(figsize=(16, 9))
plt.bar(data.index, data.values, color=NETFLIX_COLOR)
plt.title("Distribution of Genres")
plt.xlabel("Genres")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.plot()
```

## Duration of Movies

```python
movie_durations = df.loc[df["type"] == "Movie", "duration"].dropna()
# Extracts the integer part from the Series
movie_durations_ints = movie_durations.str.extract(r"(\d+)").astype(int)

df["movie_duration"] = movie_durations_ints

mean_duration = np.mean(movie_durations_ints)

plt.hist(movie_durations_ints, bins=int(np.sqrt(len(movie_durations_ints))), 
         label=f"Avg. duration: {mean_duration:.2f} mins", color=NETFLIX_COLOR )
plt.axvline(mean_duration, color="#333")
plt.legend()
plt.title("Duration of Movies")
plt.xlabel("Time (mins)")
plt.plot()
```

- The spike around 30-60 mins refers to documentaries but these are catergorised as "Movies"
- Movies with duration over 250 mins tend to have incorrect duration times stored, with the acutal duration being much lower
